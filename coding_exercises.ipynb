+ Here’s a beginner-friendly example that introduces classes, objects, and methods using the classic “Hello, world!” idea:

```python

# Define a class
class Greeter:
    # Method inside the class
    def say_hello(self):
        print("Hello, world!")

# Create an object (instance of the class)
g = Greeter()

# Call the method using the object
g.say_hello()
```

Key Concepts:

Class = Blueprint for objects.
Object = Actual thing created from the class.
Method = Function that belongs to a class and usually acts on its object.

+ This output is the same as before, can we have something different?

```python

# Define a class with a constructor and a method
class Greeter:
    def __init__(self, name):
        # Store the name in the object
        self.name = name

    def say_hello(self):
        # Use the stored name in the greeting
        print(f"Hello, {self.name}!")

# Create objects with different names
g1 = Greeter("Alice")
g2 = Greeter("Bob")

# Call the method for each object
g1.say_hello()
g2.say_hello()
```

Step-by-Step Explanation:
1. __init__ Method

This is called the constructor. It runs automatically when you create an object. self.name = name stores the name inside the object.

2. say_hello Method

Uses an f-string (f"Hello, {self.name}!") to insert the name dynamically. So the greeting changes based on the object’s data.

3. Creating Objects

Greeter("Alice") creates an object with name = "Alice". Greeter("Bob") creates another object with name = "Bob".

## Code

```python

# Sales data from your example
sales_data = ["hamburger", "hamburger", "fries", "hamburger", "fries", "doners", "doners", "ice cream"]

# Step 1: Create an empty dictionary to store counts
item_counts = {}

# Step 2: Count occurrences of each item
for item in sales_data:
    if item in item_counts:
        item_counts[item] += 1
    else:
        item_counts[item] = 1

# Step 3: Find the item with the highest count
most_popular_item = None
highest_count = 0

for item, count in item_counts.items():
    if count > highest_count:
        highest_count = count
        most_popular_item = item

# Step 4: Print the result
print("Most popular menu item is:", most_popular_item)
```

# OK, now let's begin to Getting Started with scikit-learn Datasets 


## 1. Working with Datasets
**Goal**  
Load built-in datasets from Scikit-Learn, understand the dataset structure (features and labels), and perform basic exploration.

**Key Concepts**  
- `datasets.load_iris()`: Loads the Iris dataset.
- Typical dataset components:
  - `data`: Feature matrix (2D array).
  - `target`: Labels (classification outcomes).
  - `feature_names`: Names of the features.

**Questions**  
- What keys does the dataset contain?
- What does each feature represent?
- How are the labels encoded?


```python

from sklearn import datasets
import pandas as pd
import matplotlib.pyplot as plt


iris = datasets.load_iris()
print(iris.keys())  


df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['target'] = iris.target
print(df.head())


plt.scatter(df['sepal length (cm)'], df['sepal width (cm)'], c=df['target'])
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.show()

```


## 2. Data Preprocessing
**Goal**  
Prepare the data by handling missing values, encoding categorical variables, and scaling features.

**Key Concepts**  
- Standardization: Transform features to have mean = 0 and variance = 1.
- Why scale features? → To prevent features with larger ranges from dominating the model.

**Questions**  
- What is the range of the data after standardization?
- In which scenarios is feature scaling essential?


```python

from sklearn.preprocessing import StandardScaler


scaler = StandardScaler()
scaled_data = scaler.fit_transform(iris.data)
print(scaled_data[:5])
```


## 3. Splitting Data
**Goal**  
Split the dataset into training and testing sets to avoid overfitting.

**Key Concepts**  
- `train_test_split()`: Common function for splitting data.
- Why split data? → To ensure the model generalizes well to unseen data.

**Questions**  
- What is the ratio of training to testing data?
- What happens if you do not split the data?


```python

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.3, random_state=42)
print(X_train.shape, X_test.shape)
```


## 4. Choosing and Training a Model
**Goal**  
Select an algorithm (e.g., KNN, Linear Regression) and train the model.

**Key Concepts**  
- Model fitting (`fit`): Learning parameters from training data.
- KNN: A distance-based classification algorithm.

**Questions**  
- Which algorithm did you choose and why?
- What can you do after the model is trained?


```python

from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier(n_neighbors=3)
model.fit(X_train, y_train)
```


## 5. Evaluating Model Performance
**Goal**  
Assess the model’s performance using metrics like accuracy and confusion matrix.

**Key Concepts**  
- `accuracy_score`: Proportion of correct predictions.
- Confusion Matrix: Shows correct and incorrect predictions for each class.

**Questions**  
- What is the model’s accuracy?
- Which classes are most frequently misclassified?


```python

from sklearn.metrics import accuracy_score, confusion_matrix

y_pred = model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```


## 1.Correlation Heatmap of Wine Features
The heatmap shows the correlation between all features in the wine dataset.  
- Dark red cells indicate strong positive correlation, while dark blue cells indicate strong negative correlation.
- For example, `alcohol` and `color_intensity` have a strong positive correlation, meaning wines with higher alcohol content tend to have higher color intensity.
- Highly correlated features can sometimes lead to multicollinearity in models, which is why dimensionality reduction (like PCA) can be useful.

## 2.PCA Visualization of Wine Dataset
This scatter plot represents the wine dataset reduced to two principal components using PCA.  
- Each point corresponds to a wine sample, colored by its target class (wine type).
- PCA helps capture the maximum variance in the data with fewer dimensions.
- We can see that the three classes are fairly well separated in this 2D space, which suggests that classification models can perform well on this dataset.


```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_wine
from sklearn.decomposition import PCA

wine = load_wine()
df_wine = pd.DataFrame(wine.data, columns=wine.feature_names)
df_wine['target'] = wine.target

plt.figure(figsize=(10, 8))
sns.heatmap(df_wine.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap of Wine Features")
plt.show()

pca = PCA(n_components=2)
wine_pca = pca.fit_transform(wine.data)

plt.figure(figsize=(8, 6))
plt.scatter(wine_pca[:, 0], wine_pca[:, 1], c=wine.target, cmap='viridis', edgecolor='k', s=50)
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.title('Wine Dataset PCA Visualization')
plt.colorbar(label='Target Class')
plt.show()
```
